```python
import pandas as pd

df = pd.read_csv('country_wise_latest.csv')
df.head()
```

```python
top = df.sort_values(by='Confirmed', ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure()
plt.bar(top['Country/Region'], top['Confirmed'])
plt.xticks(rotation=45)
plt.title("Top 10 Countries - COVID Cases")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Confirmed'], df['Deaths'])
plt.title("Deaths vs Confirmed Cases")
plt.xlabel("Confirmed")
plt.ylabel("Deaths")
plt.show()
```

```python
top5 = df.sort_values(by='Confirmed', ascending=False).head(5)

plt.figure()
plt.pie(top5['Confirmed'], labels=top5['Country/Region'], autopct='%1.1f%%')
plt.title("Top 5 Countries Share")
plt.show()
```

```python
# Check missing values
df.isnull().sum()

# Fill missing values
df.fillna(0, inplace=True)

# Check duplicates
df.drop_duplicates(inplace=True)
```

```python
df['Death Rate (%)'] = (df['Deaths'] / df['Confirmed']) * 100
```

```python
top = df.sort_values(by='Death Rate (%)', ascending=False).head(10)

plt.figure()
plt.barh(top['Country/Region'], top['Death Rate (%)'])
plt.title("Top 10 Countries by Death Rate")
plt.xlabel("Death Rate (%)")
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.heatmap(df[['Confirmed','Deaths','Recovered','Active']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
top_active = df.sort_values(by='Active', ascending=False).head(10)

plt.figure()
plt.bar(top_active['Country/Region'], top_active['Active'])
plt.xticks(rotation=45)
plt.title("Top 10 Active Cases")
plt.show()
```

```python
n = 7  # change this anytime

top_n = df.sort_values(by='Confirmed', ascending=False).head(n)

plt.figure()
plt.bar(top_n['Country/Region'], top_n['Confirmed'])
plt.xticks(rotation=45)
plt.title(f"Top {n} Countries by Confirmed Cases")
plt.show()
```

```python
df['Recovery Rate (%)'] = (df['Recovered'] / df['Confirmed']) * 100
```

```python
plt.figure()
plt.scatter(df['Recovery Rate (%)'], df['Death Rate (%)'])
plt.xlabel("Recovery Rate (%)")
plt.ylabel("Death Rate (%)")
plt.title("Recovery vs Death Rate")
plt.show()
```

```python
high_recovery = df.sort_values(by='Recovery Rate (%)', ascending=False).head(5)

print("Top 5 Countries with Highest Recovery Rate:")
print(high_recovery[['Country/Region','Recovery Rate (%)']])
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.bar(top_n['Country/Region'], top_n['Confirmed'])
plt.xticks(rotation=45)
plt.title("Top Cases")

plt.subplot(1,2,2)
plt.bar(top_n['Country/Region'], top_n['Deaths'])
plt.xticks(rotation=45)
plt.title("Top Deaths")

plt.show()
```
